### ¿Que se prodría hacer en casos de tener un valor nulo?

¿Cómo lidiar con los datos faltantes? En nuestro dataFrame hay algunos casos donde no tenemos datos, como por ejemplo en las columnas 'fecha_de_pago', 'cantidad_de_pagos' y 'género'. Obviamente todos estos datos faltantes podrían afectar un análisis de datos o la creación de modelos de aprendizaje de máquina. Por tanto tenemos que lidiar y aplicar técnicas para borrarlos o para completarlos.

Hay diversas maneras de manejar los datos faltantes, como por ejemplo:

- **Dropearlos:** Podemos sacar o eliminar los datos nulos con el método ```dropna()```. Se recomienda exclusivamente en casos donde contenemos mucha cantidad de registros válidos, un proceso de limpieza común es quitar cualquier inconsistencia que nos podremos encontrar, ya sea en general o en columnas específicas.


- **Reemplazarlos con valores por defecto:** Podemos definir nuestros valores por defecto dentro de un diccionario, especificamos nombre de la columna y su valor por defecto. Por ejemplo: ```valores_por_defecto = {'nombre' : 'DESCONOCIDO', ‘edad’ : 18}```.


- **Reemplazarlos con el promedio:** Podemos llamar al método ```mean()``` para que automáticamente calcule el promedio de los datos y hacer la imputación con el mismo.


- **Reemplazarlos con la media:** Podemos llamar al método ```median()``` para que automáticamente calcule la media, que es el valor que está en medio de todos los valores y hacer la imputación con el mismo.


- **Reemplazarlos con el valor más repetido (moda):** La moda es el valor más común que se encuentra en nuestra columna, lo llamamos con el método ```mode()```. Hay casos que nuestra moda puede consistir en varias, pueden tener varios valores que empatan. Por ejemplo; pueden haber 50 personas que tengan 22 años y también 50 personas que tengan 45. Ambos serían los números que se repiten más. Por lo tanto tendríamos una serie y nos regresaría un tipo vector de modas. Por tal razon, nosotros tenemos que especificar cual moda queremos que nos regrese, y lo hacemos agregando un 0 entre corchetes al final de tal manera: ```df.cantidad_de_pagos = df.cantidad_de_pagos.fillna(df.cantidad_de_pagos.mode()[0])```  (agregamos un 0 porque en python el primer numero es 0 y no 1).


- **Reemplazarlos con el valor de la celda siguiente o anterior:** Podemos utilizar los métodos ```bfill``` para imputar el dato de la celda siguiente o ```ffill``` de la anterior.

Observamos que la columna 'fecha_de_pago' es de tipo object y no datetime, 'codigo_alumno', 'codigo_carrera' y 'periodo' son formato int pero debe de ser string ya que luego no queremos hacer algún tipo de cálculo con estos datos. Por otro lado, 'materias_aprobadas' y 'cantidad_de_pagos' deben ser int, ya que no contienen decimales.

Procedemos a realizar dichos cambios:

Hacemos uso de la sentencia ```pd.to_datetime``` para realizar el cambio de 'fecha_de_pago'. (Hacemos la mudanza desde la tabla original y no de una copia para no tener errores ni advertencias)

````python
df.fecha_de_pago = pd.to_datetime(df.fecha_de_pago)
````

Hacemos uso de la sentencia ```astype()``` para realizar el cambio de 'codigo_alumno', 'codigo_carrera' y 'periodo' a str.

````python
df.codigo_alumno = df.codigo_alumno.astype(str)
df.codigo_carrera = df.codigo_carrera.astype(str)
df.periodo = df.periodo.astype(str)
````

También hacemos uso de la sentencia astype() para realizar el cambio de 'materias_aprobadas' y 'cantidad_de_pagos' a int.

````python
df.materias_aprobadas = df.materias_aprobadas.astype(int)
df.cantidad_de_pagos = df.cantidad_de_pagos.astype(int)
````

Observamos que los cambios se realizaron con exito y ahora son de los tipos correspondientes.

````python
df.dtypes
````
